```python
import numpy as np
import scipy.signal as signal
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'C:\Users\IP SLIM 3i\Music\Not Like Us - wav.wav'

# Path folder untuk menyimpan hasil
output_folder = r'C:\Users\IP SLIM 3i\Desktop\belajar'

# Cek apakah folder output ada
if not os.path.exists(output_folder):
    print(f"Folder tidak ditemukan: {output_folder}")
else:
    # Cek apakah file audio ada
    if not os.path.exists(file_path):
        print(f"File tidak ditemukan: {file_path}")
    else:
        # Muat file audio
        y, sr = librosa.load(file_path, sr=None)

        # Desain band-pass FIR filter dengan jendela rectangular
        numtaps = 101  # Jumlah tap filter
        lowcut = 500  # Frekuensi cut-off bawah (Hz)
        highcut = 1500  # Frekuensi cut-off atas (Hz)
        fir_coeff = signal.firwin(numtaps, [lowcut, highcut], pass_zero=False, window='boxcar', fs=sr)

        # Terapkan filter FIR pada sinyal
        filtered_signal_fir = signal.lfilter(fir_coeff, 1.0, y)

        # Path file output
        output_path = os.path.join(output_folder, 'filtered_audio_bandpass_fir.wav')
        sf.write(output_path, filtered_signal_fir, sr)

        # Plot sinyal asli dan sinyal hasil filter
        plt.figure(figsize=(15, 12))

        plt.subplot(4, 1, 1)
        plt.plot(y)
        plt.title('Sinyal Asli')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        plt.subplot(4, 1, 2)
        plt.plot(filtered_signal_fir)
        plt.title('Sinyal Hasil Band-pass FIR Filter dengan Jendela Rectangular')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        # Plot impuls respons
        plt.subplot(4, 1, 3)
        plt.stem(fir_coeff, use_line_collection=True)
        plt.title('Impuls Respons dari Band-pass FIR Filter')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        # Plot frekuensi respons
        plt.subplot(4, 1, 4)
        w, h = signal.freqz(fir_coeff, worN=8000)
        plt.plot(0.5 * sr * w / np.pi, np.abs(h), 'b')
        plt.title('Frekuensi Respons dari Band-pass FIR Filter')
        plt.xlabel('Frekuensi (Hz)')
        plt.ylabel('Gain')
        plt.grid()

        plt.tight_layout()
        plt.show()

        print(f"Sinyal hasil filter disimpan di: {output_path}")
```

```python
import numpy as np
import scipy.signal as signal
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'C:\Users\IP SLIM 3i\Music\Not Like Us - wav.wav'

# Path folder untuk menyimpan hasil
output_folder = r'C:\Users\IP SLIM 3i\Desktop\belajar'

# Cek apakah folder output ada
if not os.path.exists(output_folder):
    print(f"Folder tidak ditemukan: {output_folder}")
else:
    # Cek apakah file audio ada
    if not os.path.exists(file_path):
        print(f"File tidak ditemukan: {file_path}")
    else:
        # Muat file audio
        y, sr = librosa.load(file_path, sr=None)

        # Desain band-pass IIR filter dengan Kaiser window
        lowcut = 500  # Frekuensi cut-off bawah (Hz)
        highcut = 1500  # Frekuensi cut-off atas (Hz)
        order = 2  # Orde filter yang lebih rendah
        N, beta = signal.kaiserord(60, 2 / sr * (highcut - lowcut))
        
        # Desain filter IIR menggunakan Kaiser window
        b, a = signal.iirfilter(N, [lowcut, highcut], rs=60, btype='band', ftype='cheby2', fs=sr, output='ba', analog=False)

        # Terapkan filter IIR pada sinyal
        filtered_signal_iir = signal.lfilter(b, a, y)

        # Path file output
        output_path = os.path.join(output_folder, 'filtered_audio_bandpass_iir_kaiser.wav')
        sf.write(output_path, filtered_signal_iir, sr)

        # Plot sinyal asli dan sinyal hasil filter
        plt.figure(figsize=(15, 12))

        plt.subplot(4, 1, 1)
        plt.plot(y)
        plt.title('Sinyal Asli')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        plt.subplot(4, 1, 2)
        plt.plot(filtered_signal_iir)
        plt.title('Sinyal Hasil Band-pass IIR Filter dengan Kaiser Window')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        # Plot impuls respons
        impulse = np.zeros(100)
        impulse[0] = 1
        impulse_response = signal.lfilter(b, a, impulse)

        plt.subplot(4, 1, 3)
        plt.stem(impulse_response, use_line_collection=True)
        plt.title('Impuls Respons dari Band-pass IIR Filter')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        # Plot frekuensi respons
        w, h = signal.freqz(b, a, worN=8000)
        plt.subplot(4, 1, 4)
        plt.plot(0.5 * sr * w / np.pi, np.abs(h), 'b')
        plt.title('Frekuensi Respons dari Band-pass IIR Filter')
        plt.xlabel('Frekuensi (Hz)')
        plt.ylabel('Gain')
        plt.grid()

        plt.tight_layout()
        plt.show()

        print(f"Sinyal hasil filter disimpan di: {output_path}")
```
