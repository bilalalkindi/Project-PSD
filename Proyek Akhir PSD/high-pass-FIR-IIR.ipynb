```python
import numpy as np
import scipy.signal as signal
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'C:\Users\IP SLIM 3i\Music\Not Like Us - wav.wav'

# Path folder untuk menyimpan hasil
output_folder = r'C:\Users\IP SLIM 3i\Desktop\belajar'

# Cek apakah folder output ada
if not os.path.exists(output_folder):
    print(f"Folder tidak ditemukan: {output_folder}")
else:
    # Cek apakah file audio ada
    if not os.path.exists(file_path):
        print(f"File tidak ditemukan: {file_path}")
    else:
        # Muat file audio
        y, sr = librosa.load(file_path, sr=None)

        # Desain high-pass FIR filter dengan jendela rectangular
        numtaps = 101  # Jumlah tap filter
        cutoff = 1000  # Frekuensi cut-off high-pass (Hz)
        fir_coeff = signal.firwin(numtaps, cutoff, pass_zero=False, window='boxcar', fs=sr)

        # Terapkan filter FIR pada sinyal
        filtered_signal_fir = signal.lfilter(fir_coeff, 1.0, y)

        # Path file output
        output_path = os.path.join(output_folder, 'filtered_audio_highpass_fir_rectangle.wav')
        sf.write(output_path, filtered_signal_fir, sr)

        # Plot sinyal asli dan sinyal hasil filter
        plt.figure(figsize=(15, 10))

        plt.subplot(3, 1, 1)
        plt.plot(y)
        plt.title('Sinyal Asli')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        plt.subplot(3, 1, 2)
        plt.plot(filtered_signal_fir)
        plt.title('Sinyal Hasil High-pass FIR Filter dengan Jendela Rectangular')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        # Plot impulse response
        plt.subplot(3, 1, 3)
        plt.stem(fir_coeff, use_line_collection=True)
        plt.title('Impulse Response of the High-pass FIR Filter')
        plt.xlabel('Tap')
        plt.ylabel('Amplitude')

        plt.tight_layout()
        plt.show()

        # Plot frequency response
        w, h = signal.freqz(fir_coeff, worN=8000)
        plt.figure(figsize=(15, 6))
        plt.plot(0.5 * sr * w / np.pi, np.abs(h), 'b')
        plt.title('Frequency Response of the High-pass FIR Filter')
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Gain')
        plt.grid()
        plt.show()

        print(f"Sinyal hasil filter disimpan di: {output_path}")
```
