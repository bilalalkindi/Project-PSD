```python
import numpy as np
import scipy.signal as signal
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'D:\KULIAH FAWWAZ\SEMESTER 4\PENGOLAHAN SINYAL DIGITAL\kelompok uas\Music\Not Like Us - wav.wav'
# Path folder untuk menyimpan hasil
output_folder = r'D:\KULIAH FAWWAZ\SEMESTER 4\PENGOLAHAN SINYAL DIGITAL\kelompok uas\Hasil'

# Cek apakah folder output ada
if not os.path.exists(output_folder):
    print(f"Folder tidak ditemukan: {output_folder}")
else:
    # Cek apakah file audio ada
    if not os.path.exists(file_path):
        print(f"File tidak ditemukan: {file_path}")
    else:
        # Muat file audio
        y, sr = librosa.load(file_path, sr=None)

        # Desain low-pass FIR filter dengan jendela rectangular
        numtaps = 101  # Jumlah tap filter
        cutoff = 1000  # Frekuensi cut-off low-pass (Hz)
        fir_coeff = signal.firwin(numtaps, cutoff, pass_zero=True, window='boxcar', fs=sr)

        # Terapkan filter FIR pada sinyal
        filtered_signal_fir = signal.lfilter(fir_coeff, 1.0, y)

        # Path file output
        output_path = os.path.join(output_folder, 'filtered_audio_lowpass_fir_rectangle.wav')
        sf.write(output_path, filtered_signal_fir, sr)

        # Plot sinyal asli dan sinyal hasil filter
        plt.figure(figsize=(15, 10))

        plt.subplot(3, 1, 1)
        plt.plot(y)
        plt.title('Sinyal Asli')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        plt.subplot(3, 1, 2)
        plt.plot(filtered_signal_fir)
        plt.title('Sinyal Hasil Low-pass FIR Filter dengan Jendela Rectangular')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        # Plot impulse response
        plt.subplot(3, 1, 3)
        plt.stem(fir_coeff)
        plt.title('Impulse Response of the Low-pass FIR Filter')
        plt.xlabel('Tap')
        plt.ylabel('Amplitude')

        plt.tight_layout()
        plt.show()

        # Plot frequency response
        w, h = signal.freqz(fir_coeff, worN=8000)
        plt.figure(figsize=(15, 6))
        plt.plot(0.5 * sr * w / np.pi, np.abs(h), 'b')
        plt.title('Frequency Response of the Low-pass FIR Filter')
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Gain')
        plt.grid()
        plt.show()

        print(f"Sinyal hasil filter disimpan di: {output_path}")
```

```python
import numpy as np
import scipy.signal as signal
import scipy.signal.windows as windows
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'D:\KULIAH FAWWAZ\SEMESTER 4\PENGOLAHAN SINYAL DIGITAL\kelompok uas\Music\Not Like Us - wav.wav'
# Path folder untuk menyimpan hasil
output_folder = r'D:\KULIAH FAWWAZ\SEMESTER 4\PENGOLAHAN SINYAL DIGITAL\kelompok uas\Hasil'
# Cek apakah folder output ada
if not os.path.exists(output_folder):
    print(f"Folder tidak ditemukan: {output_folder}")
else:
    # Cek apakah file audio ada
    if not os.path.exists(file_path):
        print(f"File tidak ditemukan: {file_path}")
    else:
        # Muat file audio
        y, sr = librosa.load(file_path, sr=None)

        # Desain low-pass IIR filter menggunakan Butterworth
        cutoff = 1000  # Frekuensi cut-off low-pass (Hz)
        nyquist = 0.5 * sr
        normal_cutoff = cutoff / nyquist

        # Dapatkan koefisien filter Butterworth
        b, a = signal.butter(N=4, Wn=normal_cutoff, btype='low', analog=False)

        # Terapkan window Kaiser pada koefisien filter Butterworth
        numtaps = len(b)  # Panjang filter
        beta = 14  # Nilai beta untuk jendela Kaiser, biasanya disarankan antara 6 dan 14
        kaiser_window = windows.kaiser(numtaps, beta)
        b = b * kaiser_window

        # Terapkan filter IIR pada sinyal
        filtered_signal_iir = signal.lfilter(b, a, y)

        # Path file output
        output_path = os.path.join(output_folder, 'filtered_audio_lowpass_iir_butterworth_kaiser.wav')
        sf.write(output_path, filtered_signal_iir, sr)

        # Plot sinyal asli dan sinyal hasil filter
        plt.figure(figsize=(15, 9))

        plt.subplot(3, 1, 1)
        plt.plot(y)
        plt.title('Sinyal Asli')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        plt.subplot(3, 1, 2)
        plt.plot(filtered_signal_iir)
        plt.title('Sinyal Hasil Low-pass IIR Filter (Butterworth + Kaiser)')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        # Plot impulse response
        plt.subplot(3, 1, 3)
        impulse = np.zeros(100)
        impulse[0] = 1
        impulse_response = signal.lfilter(b, a, impulse)
        plt.stem(impulse_response)
        plt.title('Impulse Response of the Low-pass IIR Filter (Butterworth + Kaiser)')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        plt.tight_layout()
        plt.show()

        # Plot frequency response
        w, h = signal.freqz(b, a, worN=8000)
        plt.figure(figsize=(10, 6))
        plt.plot(0.5 * sr * w / np.pi, np.abs(h), 'b')
        plt.title('Frequency Response of the Low-pass IIR Filter (Butterworth + Kaiser)')
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Gain')
        plt.grid()
        plt.show()

        print(f"Sinyal hasil filter disimpan di: {output_path}")
```

```python
import numpy as np
import scipy.signal as signal
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'D:\KULIAH FAWWAZ\SEMESTER 4\PENGOLAHAN SINYAL DIGITAL\kelompok uas\Music\Not Like Us - wav.wav'
# Path folder untuk menyimpan hasil
output_folder = r'D:\KULIAH FAWWAZ\SEMESTER 4\PENGOLAHAN SINYAL DIGITAL\kelompok uas\Hasil'
# Cek apakah folder output ada
if not os.path.exists(output_folder):
    print(f"Folder tidak ditemukan: {output_folder}")
else:
    # Cek apakah file audio ada
    if not os.path.exists(file_path):
        print(f"File tidak ditemukan: {file_path}")
    else:
        # Muat file audio
        y, sr = librosa.load(file_path, sr=None)

        # Desain low-pass IIR filter menggunakan Butterworth
        cutoff = 1000  # Frekuensi cut-off low-pass (Hz)
        nyquist = 0.5 * sr
        normal_cutoff = cutoff / nyquist

        # Dapatkan koefisien filter Butterworth
        b, a = signal.butter(N=4, Wn=normal_cutoff, btype='low', analog=False)

        # Terapkan window Gaussian pada koefisien filter Butterworth
        numtaps = len(b)  # Panjang filter
        sigma = 0.5 * numtaps  # Sigma untuk jendela Gaussian
        gaussian_window = np.exp(-0.5 * (np.arange(numtaps) - (numtaps - 1) / 2)**2 / sigma**2)
        b *= gaussian_window

        # Terapkan filter IIR pada sinyal
        filtered_signal_iir = signal.lfilter(b, a, y)

        # Path file output
        output_path = os.path.join(output_folder, 'filtered_audio_lowpass_iir_butterworth_gaussian.wav')
        sf.write(output_path, filtered_signal_iir, sr)

        # Plot sinyal asli dan sinyal hasil filter
        plt.figure(figsize=(15, 9))

        plt.subplot(3, 1, 1)
        plt.plot(y)
        plt.title('Sinyal Asli')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        plt.subplot(3, 1, 2)
        plt.plot(filtered_signal_iir)
        plt.title('Sinyal Hasil Low-pass IIR Filter (Butterworth + Gaussian)')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        # Plot impulse response
        plt.subplot(3, 1, 3)
        impulse_response = signal.lfilter(b, a, [1] + [0] * (len(b) - 1))
        plt.stem(impulse_response)
        plt.title('Impulse Response of the Low-pass IIR Filter (Butterworth + Gaussian)')
        plt.xlabel('Sample')
        plt.ylabel('Amplitude')

        plt.tight_layout()
        plt.show()

        # Plot frequency response
        w, h = signal.freqz(b, a, worN=8000)
        plt.figure(figsize=(10, 6))
        plt.plot(0.5 * sr * w / np.pi, np.abs(h), 'b')
        plt.title('Frequency Response of the Low-pass IIR Filter (Butterworth + Gaussian)')
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Gain')
        plt.grid()
        plt.show()

        print(f"Sinyal hasil filter disimpan di: {output_path}")
```

```python
import numpy as np
import scipy.signal as signal
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import os

# Path file audio input
file_path = r'D:\KULIAH FAWWAZ\SEMESTER 4\PENGOLAHAN SINYAL DIGITAL\kelompok uas\Music\Not Like Us - wav.wav'
# Path folder untuk menyimpan hasil
output_folder = r'D:\KULIAH FAWWAZ\SEMESTER 4\PENGOLAHAN SINYAL DIGITAL\kelompok uas\Hasil'

# Cek apakah folder output ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)  # Buat folder jika belum ada

# Cek apakah file audio ada
if not os.path.exists(file_path):
    print(f"File tidak ditemukan: {file_path}")
else:
    # Muat file audio
    y, sr = librosa.load(file_path, sr=None)

    # Desain low-pass IIR filter menggunakan Butterworth
    cutoff = 1000  # Frekuensi cut-off low-pass (Hz)
    nyquist = 0.5 * sr
    normal_cutoff = cutoff / nyquist

    # Dapatkan koefisien filter Butterworth
    b, a = signal.butter(N=4, Wn=normal_cutoff, btype='low', analog=False)

    # Terapkan window Tukey pada koefisien filter Butterworth
    numtaps = len(b)  # Panjang filter
    alpha = 0.5  # Parameter alpha untuk jendela Tukey, 0 untuk jendela rectangular, 1 untuk jendela Hann
    tukey_window = 0.5 * (1 + np.cos(2 * np.pi * np.arange(numtaps) / (numtaps - 1) - 1))  # Jendela Tukey secara manual
    b *= tukey_window

    # Buat respons impuls secara manual
    impulse = np.zeros_like(b)
    impulse[0] = 1
    impulse_response = signal.lfilter(b, a, impulse)

    # Path file output
    output_path = os.path.join(output_folder, 'filtered_audio_lowpass_iir_butterworth_tukey.wav')
    sf.write(output_path, impulse_response, sr)

    # Plot sinyal asli dan sinyal hasil filter
    plt.figure(figsize=(15, 9))

    plt.subplot(3, 1, 1)
    plt.plot(y)
    plt.title('Sinyal Asli')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')

    plt.subplot(3, 1, 2)
    plt.plot(impulse_response)
    plt.title('Sinyal Hasil Low-pass IIR Filter (Butterworth + Tukey)')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')

    # Plot impulse response
    plt.subplot(3, 1, 3)
    plt.stem(impulse_response)
    plt.title('Impulse Response of the Low-pass IIR Filter (Butterworth + Tukey)')
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')

    plt.tight_layout()
    plt.show()

    # Plot frequency response
    w, h = signal.freqz(b, a, worN=8000)
    plt.figure(figsize=(10, 6))
    plt.plot(0.5 * sr * w / np.pi, np.abs(h), 'b')
    plt.title('Frequency Response of the Low-pass IIR Filter (Butterworth + Tukey)')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Gain')
    plt.grid()
    plt.show()

    print(f"Sinyal hasil filter disimpan di: {output_path}")
```
