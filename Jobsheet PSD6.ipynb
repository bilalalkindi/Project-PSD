# IIR VS FIR

### Sinyal dan Filter

Pada bagian sebelumnya, kami memperkenalkan konsep sinyal, spektrum, dan bentuk Z. Selama perjalanan sinyal dari sumber ke penerima, urutan data mengalami perubahan dan transformasi. Transformasi ini terjadi saat sinyal melewati berbagai perangkat komputasi yang melakukan operasi matematika tertentu. Dalam pemrosesan sinyal digital, data melewati rangkaian digital yang disebut filter. Filter digital, mirip dengan filter analog, memiliki karakteristik yang berbeda. Biasanya, karakteristik ini dinyatakan dalam representasi frekuensi - yaitu, karakteristik amplitudo-frekuensi dan fase-frekuensi dari filter. Filter digital terutama digunakan untuk meningkatkan kualitas sinyal - baik untuk mengekstraksi sinyal dari urutan data, atau untuk mengurangi sinyal yang tidak diinginkan - dengan menekan sinyal tertentu dalam urutan sampel yang diterima.

Saat ini, filter digital digunakan hampir di mana saja di mana pemrosesan sinyal diperlukan. Bahkan operasi matematika dasar (perkalian dan penjumlahan) dapat direpresentasikan sebagai filter digital. Oleh karena itu, semua perangkat di sekitar kita (komputer, telepon, televisi, dll.) mengandung berbagai filter digital. Misalnya, televisi menggunakan filter untuk mengubah tampilan gambar. Penyesuaian kecerahan pada dasarnya adalah proses filtrasi gambar. Filter digital digunakan dalam berbagai tugas pemrosesan sinyal, seperti analisis spektral, pemrosesan data audio dan video, pemrosesan ucapan, dan pemrosesan gerakan. Khusus untuk tugas-tugas DSP, filter adalah sistem matematis yang mengubah bentuk sinyal masukan dalam hal amplitudo, fase, dan frekuensi.

Keunggulan Filter Digital:

- Kemampuan untuk mencapai karakteristik yang tidak mungkin diperoleh dengan metode analog, seperti respons frekuensi yang tepat linear.
- Filter digital lebih stabil terhadap perubahan lingkungan eksternal (suhu, kelembaban, tekanan), sehingga tidak memerlukan kalibrasi yang terus-menerus.
- Kemudahan dalam mengubah frekuensi pemrosesan sinyal.
- Tergantung pada kecepatan pemrosesan dan frekuensi sinyal, satu filter dapat memproses data dari beberapa saluran secara bersamaan.
- Data sebelum dan setelah filter dapat disimpan dalam media penyimpanan apa pun, sementara sinyal analog sulit untuk direkam tanpa konversi ke bentuk digital.
- Konsumsi daya rendah dan biaya lebih rendah dibandingkan dengan filter analog.
- Tingkat akurasi yang lebih tinggi dalam perhitungan, karena perangkat digital modern tidak terbatas oleh akurasi perhitungan.
- Kemudahan dalam merancang filter dengan karakteristik frekuensi yang kompleks.
- Dalam tugas-tugas biomedis, penggunaan filter analog pada frekuensi yang sangat rendah sulit, berbeda dengan filter digital.


Kekurangan Filter Digital:

- Keterbatasan resolusi: Dalam perhitungan pada filter digital dengan resolusi terbatas, terjadi akumulasi noise kuantisasi dan noise pembulatan. Salah satu tugas umum dalam DSP adalah memilih resolusi filter yang sesuai.
- Keterbatasan kecepatan pemrosesan: Biasanya, filter analog mampu memproses data pada frekuensi yang sangat tinggi, sementara filter digital dibatasi oleh frekuensi sampling.
- Keterbatasan perangkat keras: Karena sinyal nyata di alam adalah kontinu, ADC (Analog-to-Digital Converter) dan DAC (Digital-to-Analog Converter) digunakan untuk memprosesnya. Kualitas sinyal digital (atau analog) yang dihasilkan juga tergantung pada karakteristik perangkat ini.

### Filter IIR dan FIR

Filter FIR (finite impulse response) adalah filter digital dengan respon impuls yang terbatas. Ini berarti bahwa karakteristik impuls filter ini dibatasi dalam waktu, memiliki sejumlah koefisien yang terhitung, dan mulai dari suatu titik waktu tertentu, responnya menjadi nol. Jika sinyal impuls satu satuan diberikan pada input filter FIR, keluaran filter akan memiliki jumlah sampel yang terbatas. Biasanya, respon frekuensi filter dengan karakteristik impuls yang terbatas adalah linear.

Secara umum, filter FIR diimplementasikan tanpa umpan balik, yang berarti mereka non-rekursif. Namun, dengan menggunakan transformasi matematika, filter ini dapat diubah menjadi bentuk rekursif.

Filter IIR (infinite impulse response) adalah filter digital dengan karakteristik impuls yang tidak terbatas dalam waktu, yang berarti memiliki jumlah koefisien yang sangat besar atau bahkan tak terbatas. Filter IIR juga dikenal sebagai filter rekursif karena menggunakan umpan balik dalam implementasinya (sinyal dari output filter melalui elemen penundaan kembali masuk ke filter dan mempengaruhi perubahan di dalamnya). Fungsi transfer filter IIR berbentuk pecahan rasional. Beberapa jenis filter IIR yang terkenal termasuk filter Chebyshev, Butterworth, Kalman, Bessel, dan lain-lain.

Fungsi transfer dapat diimplementasikan dengan berbagai skema struktural. Skema struktural yang paling umum adalah skema serial dan paralel. Skema serial biasanya digunakan ketika fungsi transfer filter rekursif dapat diuraikan menjadi produk fungsi transfer orde pertama dan kedua. Sebaliknya, skema paralel digunakan ketika fungsi transfer filter rekursif dapat diuraikan menjadi jumlah fungsi transfer orde pertama dan kedua.

</math>, sementara penjumlah menggabungkan hasil yang diperoleh.

Skema struktural dari filter FIR non-rekursif sesuai dengan persamaan:


Jumlah elemen penunda selalu satu kurang dari jumlah koefisien atau orde filter. Karakteristik impuls filter FIR sesuai dengan kumpulan koefisien dari filter tersebut.

Skema struktural dari filter IIR terlihat lebih kompleks karena adanya umpan balik.

Jumlah elemen penunda pada filter IIR adalah N + M

Skema struktural dalam bentuk kanonik mengurangi jumlah elemen penunda dengan menggunakan elemen penunda yang bersamaan. Penjumlah masukan menggunakan umpan balik. Penjumlah keluaran menggabungkan hasil. Transformasi dari satu bentuk ke bentuk lainnya cukup mudah dan sederhana.

### Pemilihan Filter

Pemilihan jenis filter sering kali menjadi pertimbangan penting bagi pengembang perangkat digital. Keputusan antara menggunakan filter FIR atau IIR didasarkan pada kriteria berikut:

- Karakteristik fase filter IIR bersifat non-linear, sedangkan filter FIR dapat memiliki karakteristik fase yang sepenuhnya linear, yang berarti tidak menyebabkan distorsi pada bentuk sinyal.
- Filter FIR stabil karena diimplementasikan dalam bentuk non-rekursif, sedangkan filter IIR dapat menjadi tidak stabil.
- Dalam aplikasi nyata, masalah resolusi data (pembulatan, kuantisasi) menjadi kurang signifikan pada perangkat tanpa umpan balik, seperti pada filter FIR.
- Untuk mencapai respons frekuensi dengan bentuk yang kompleks atau persegi maksimal, filter FIR membutuhkan jumlah koefisien yang signifikan. Filter IIR lebih efisien dalam menangani tugas ini.
- Filter FIR umumnya tidak memiliki padanan analog.
- Filter FIR memudahkan perolehan karakteristik yang diinginkan (tingkat penurunan, ketidakrataan di jalur lewat, frekuensi cut-off, dll.).
- Filter IIR jauh lebih efisien dalam hal jumlah operasi perkalian, penambahan, dan jumlah elemen penundaannya.

### Filter Butterworth

Contoh implementasi filter Butterworth orde 3. Kami akan menunjukkan bagaimana sinyal yang bising melewati filter tersebut:

```python
# Create input signal
t = 2 * np.pi * np.linspace(-1, 1, 500)
x = np.sin(0.25*t*t)+ 0.95*np.sin(2.0*t)

# Add some white noise
np.random.seed(1)
xn = x + np.random.randn(len(t))

# 3-order lowpass butterworth filter
b, a = butter(3, 0.2)
z = lfilter(b, a, xn)

wn = [0.01, 0.05, 0.1, 0.2]

# Calculate IIR filter
zz = np.zeros((t.size, 4))
for i in range(4):
    b, a = butter(3, wn[i])
    zz[:, i] = lfilter(b, a, xn)

# Plot results
plt.figure(figsize=(16, 8), dpi=80)
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.plot(t, xn, 'C0--', linewidth=1.5)
    plt.plot(t, zz[:,i], 'C1', linewidth=2.5)
    plt.xlim([-2 * np.pi, 2 * np.pi])
    plt.grid(True)
    plt.legend(('Signal', 'Filtered, wn = {}'.format(wn[i])), loc='lower right')
```

### Filter FIR Homogen

### Perhitungan Koefisien Filter

Terdapat berbagai metode untuk menghasilkan koefisien filter, seperti metode pemutaran jendela, sampling frekuensi, dan metode optimal seperti Chebyshev yang menggunakan algoritma Remez. Setiap metode memiliki karakteristiknya sendiri dan menghasilkan hasil yang berbeda. Metode pemutaran jendela, misalnya, dapat menghasilkan fenomena Gibbs yang mengakibatkan ketidakteraturan dalam karakteristik frekuensi filter. Fenomena ini bisa diatasi dengan memberikan toleransi terhadap ketidakteraturan dalam jalur transmisi dan pita penolakan.

Salah satu metode utama untuk menghitung koefisien filter adalah algoritma Parks-McClellan, yang juga dikenal sebagai algoritma Remez yang dimodifikasi. Algoritma ini adalah metode iteratif tak langsung yang mencari nilai-nilai optimal dengan karakteristik Chebyshev filter. Keunikan dari metode ini adalah minimisasi kesalahan dalam jalur transmisi dan pita penolakan dengan pendekatan Chebyshev terhadap karakteristik impuls.

Secara umum, semakin banyak koefisien yang digunakan, semakin sedikit ketidakteraturan dalam karakteristik frekuensi, dan semakin mendekati bentuk persegi dalam jalur transmisi. Meskipun pilihan metode dapat memengaruhi hasil akhir, tujuannya adalah sama, yaitu untuk meminimalkan lonjakan dalam jalur transmisi dan meningkatkan ketepatan karakteristik frekuensi (dengan mempertahankan keadaan linearitas fase).

### Analisis Konsekuensi dari Ketepatan Jumlah Bit yang Terbatas

## Pasangan Filter Konvulasi Tak Hingga di Python

Implementasi perangkat lunak dari filter konvolusi tak hingga melibatkan pencarian koefisien dari karakteristik impuls.

Penghitungan filter konvolusi tak hingga dapat dilakukan dengan menggunakan beberapa fungsi dari paket scipy.signal. Berikut adalah semua metode perancangan filter konvolusi tak hingga dan tinjauan detail untuk setiap metode.

### Perhitungan Filter Konvulasi Tak Hingga dengan Metode Kuadrat Terkecil 

Dengan menggunakan metode kuadrat terkecil, fungsi ini menghitung koefisien filter yang memberikan pendekatan terbaik terhadap karakteristik frekuensi yang diinginkan. Secara matematis, dalam mencari koefisien, integral dari rata-rata kuadrat error yang ditimbang dalam rentang bandwidth akan diminimalkan.

Metode firls mengembalikan koefisien filter FIR. Argumen input dari fungsi ini meliputi:
- **numtaps**: orde filter,
- **bands**: urutan monoton tidak menurun dari frekuensi. Semua elemen harus non-negatif dan kurang dari atau sama dengan frekuensi Nyquist, yang ditentukan oleh parameter nyq,
- **desired**: urutan amplitudo, berisi penguatan dan penurunan yang diinginkan pada nilai frekuensi yang ditentukan dalam bands,
- **weight**: "bobot" relatif dari bandwidth dan penekanan. Menentukan kompromi kualitas di bandwidth-bandwidth tersebut,
- **nyq atau fs**: Frekuensi Nyquist. Jika tidak ditentukan, defaultnya adalah satu.

## Perhitungan Filter Konvulasi Tak Hingga dengan Algoritma Remez

Metode remez mengembalikan koefisien filter FIR. Argumen inputnya identik dengan metode sebelumnya:

- **numtaps**: orde filter,
- **bands**: urutan monoton tidak menurun dari frekuensi. Semua elemen harus non-negatif dan kurang dari atau sama dengan frekuensi Nyquist, yang ditentukan oleh parameter Hz,
- **desired**: urutan amplitudo, berisi penguatan dan penurunan yang diinginkan pada nilai frekuensi yang ditentukan dalam bands,
- **weight**: "bobot" relatif dari bandwidth dan penekanan. Menentukan kompromi kualitas di bandwidth-bandwidth tersebut,
- **Hz atau fs**: Frekuensi Nyquist. Jika tidak ditentukan, defaultnya adalah satu.

## Perhitungan Filter Konvulasi Tak Hingga dengan Metode Jendela

Metode firwin dan firwin2 mengembalikan koefisien filter FIR.

### Metode Firwin

- **numtaps**: urutan filter,
- **cutoff**: frekuensi cutoff relatif terhadap laju sampel fs,
- **width**: lebar transisi dari jalur lintasan ke jalur penekanan (jika ditentukan),
- **window**: fungsi jendela, dapat digunakan bersama metode get_window atau ditetapkan langsung dengan nama jendela,
- **fs**: laju sampel atau frekuensi Nyquist, dengan default nilai satu jika tidak ditentukan.

### Metode Firwin 2

- **numtaps**: urutan filter,
- **freq**: urutan frekuensi monotonik tak menurun. Semua elemen harus non-negatif dan tidak melebihi frekuensi Nyquist yang ditentukan oleh parameter fs,
- **gain**: urutan amplitudo yang mencakup penguatan dan redaman yang diinginkan pada nilai frekuensi yang ditentukan dalam freq,
- **window**: fungsi jendela, dapat digunakan bersama metode get_window atau ditetapkan langsung dengan nama jendela,
- **fs**: laju sampel atau frekuensi Nyquist, dengan default nilai satu jika tidak ditentukan.

### Fungsi Jendela

Bab khusus membahas fungsi jendela dan karakteristiknya, namun beberapa fungsi jendela yang paling umum digunakan dalam praktik meliputi: Kaiser, Gauss, Blackman-Harris, jendela dengan puncak datar, Hamming, Hann, dan lain-lain.

Untuk memilih fungsi jendela, metode get_window digunakan. Argumen masukan meliputi nama jendela dan panjang karakteristik impuls.

Fungsi ini mengembalikan array satu dimensi koefisien - nilai karakteristik impuls. Koefisien ini digunakan dalam operasi konvolusi untuk membentuk karakteristik impuls filter tersebut.

### Gelombang Grup

Ini adalah fungsi bantuan untuk menghitung penundaan grup filter digital. Parameter masukan - larik koefisien a, b. Metode mengembalikan kumpulan frekuensi dan larik penundaan grup yang sesuai untuk frekuensi tersebut.

### Contoh

Contoh pertama - menggunakan fungsi firwin2. Mari tentukan larik frekuensi yang dinormalisasi dari 0 hingga 1. Tentukan penguatan pada frekuensi yang dipilih, serta urutan filter. Mari kita buat karakteristik impuls.

```python
N = 100
lst_freqs = np.linspace(0, 1, N)

lst_gain = np.zeros(N)
lst_gain[0:20] = 1

NFIR = 64
taps = firwin2(NFIR, lst_freqs, lst_gain)

plt.figure(figsize=(16, 4), dpi=100)
plt.title('Impulse responce')
plt.stem(taps, basefmt='C0')
plt.xlim([0, NFIR-1])
plt.xlabel('Samples')
plt.grid(True)
```

Mari kita buat karakteristik amplitudo-frekuensi (AF) dari filter pita dan filter turun rendah menggunakan tiga metode: firls(), remez(), dan firwin2().

```python
fs = 10
N = 117
desired = (0, 0, 1, 1, 0, 0)
bands = (0, 1, 2, 4, 4.5, 5)

fir_firls = firls(N, bands, desired, fs=fs)
fir_remez = remez(N, bands, desired[::2], fs=fs)
fir_firwin2 = firwin2(N, bands, desired, fs=fs)

plt.figure(figsize=(16, 5))
plt.title('Frequency responce')
for fir in (fir_firls, fir_remez, fir_firwin2):
    freq, resp = freqz(fir)
    resp = np.abs(resp)
    resp /= np.max(resp) + 10**(-15)
    plt.plot(freq, 20*np.log10(resp))
plt.xlim([0, np.pi])
plt.ylim([-180, 5])
plt.legend(['firls', 'remez', 'firwin2'], loc ='upper left')
plt.grid(True)
```

### Kaiser FIR Filter

Mari kita lihat bagaimana sinyal harmonik yang berisik melewati filter FIR. Kita akan menambahkan beberapa sinyal harmonik frekuensi tinggi dengan amplitudo yang berbeda ke dalam sinyal ini.

Seperti yang terlihat, sinyal berisik sulit untuk dianalisis dalam domain waktu. Namun, dengan melewatkannya melalui filter FIR dengan frekuensi cutoff yang sudah ditentukan, kita dapat memisahkan sinyal yang diinginkan dari gangguan.

Perlu dicatat bahwa semua sinyal dan noise yang masuk ke dalam jalur pass filter akan tetap ada dan akan mempengaruhi kualitas sinyal keluaran.

```python
N = 1024
t = np.linspace(0, 1, N, endpoint=True)
x = 3*np.cos(2*np.pi*20*t) + 210*np.sin(2*np.pi*117*t) + 100*np.sin(2*np.pi*184*t) + 380*np.sin(2*np.pi*281*t)

np.random.seed(1)
xn = x + 5*np.random.randn(N)

taps = 160
h = firwin(taps, 0.1, window=('kaiser', 9))
y = lfilter(h, 1, xn)

lst_sig = [xn, h, y]

sig_titles = ['Noisy signal', 'FIR impulse responce', 'Filtered signal']
fft_titles = ['Input spectrum', 'FIR filter', 'Result']


plt.figure(figsize=(16, 10))
for i in range(3):
    clc_fft = np.abs(fft(lst_sig[i], N)) 
    clc_fft = 20*np.log10(10e-11+clc_fft/np.max(clc_fft))
    
    plt.subplot(3, 2, 2*i+1)
    plt.plot(lst_sig[i], color='C'+str(i))
    plt.title(sig_titles[i])
    if (i == 2):
        plt.ylim([-5, 5])
        plt.xlim([taps, lst_sig[i].size//2-1])
    else:
        plt.xlim([0, lst_sig[i].size-1])
    plt.grid(True)
    
    plt.subplot(3, 2, 2*(i+1))
    plt.plot(clc_fft, color='C'+str(i))
    plt.title(fft_titles[i])
    plt.xlim([0, N//2-1])
    plt.grid(True)
plt.tight_layout()
```

Pada gambar di atas, terlihat bahwa sinyal keluaran yang difilter dengan sedikit distorsi menyerupai sinyal harmonik yang diinginkan. Namun, perlu dicatat bahwa mencapai bentuk gelombang harmonik yang ideal melalui filter FIR atau IIR tidaklah mungkin, karena fungsi utama filter adalah untuk memotong komponen spektral yang tidak diinginkan. Untuk meningkatkan kualitas sinyal yang difilter, metode pemrosesan sinyal digital lainnya, seperti spektral dan korelasi, dapat digunakan.
